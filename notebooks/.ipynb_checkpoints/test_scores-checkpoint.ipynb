### Splitting the data

```python
df.columns
```

```python
X = pd.get_dummies(df.drop(['posttest', 'n_student', 'student_id'], axis=1), 
                   drop_first=True)
y = df['posttest']
```

```python
len(X.columns)
```

```python
X['pretest']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=124)
```

```python
len(X_train), len(y_train), len(X_test), len(y_test)
```

### A single decision tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor()
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
len(y_pred), len(y_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
np.mean(y_test)
```

```python
mean_absolute_error(y_test, y_pred)/np.mean(y_test)
```

```python
imp_feat = dt_model.feature_importances_
```

```python
df_feat = pd.DataFrame(index=X.columns, data=dt_model.feature_importances_, 
                       columns=['features'])
```

```python
df_feat = df_feat[df_feat['features']>=0.001]
```

```python
df_feat
```

### Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model = LinearRegression()
```

```python
lin_model.fit(X_train, y_train)
```

```python
y_pred_lin = lin_model.predict(X_test)
```

```python
len(y_test), len(y_pred_lin)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred_lin))
```

```python
mean_absolute_error(y_test, y_pred_lin)
```

```python
np.mean(y_test)
```

```python
mean_absolute_error(y_test, y_pred_lin)/np.mean(y_test)
```

```python
res = y_test - y_pred_lin
```

```python
sns.histplot(res);
```

```python
sns.scatterplot(x=y_test, y=res)
plt.axhline(y=0, color='red', linestyle='--')
plt.ylabel('residual')
plt.show()
```

```python
lin_model.coef_
```

```python
df_coef_lin = pd.DataFrame(index=X.columns, 
                          data=lin_model.coef_, 
                          columns=['coefficients'])
```

```python
len(df_coef_lin), len(X), len(X.columns)
```

### A random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor()
```

```python
rf_model.fit(X_train.drop('pretest', axis=1), y_train)
```

```python
y_pred_rf = rf_model.predict(X_test.drop('pretest', axis=1))
```

```python
np.sqrt(mean_squared_error(y_test, y_pred_rf))
```

```python
mean_absolute_error(y_test, y_pred_rf)
```

```python
mean_absolute_error(y_test, y_pred_rf)/np.mean(y_test)
```

```python
imp_feat_rf = rf_model.feature_importances_
```

```python
df_feat_rf = pd.DataFrame(index=X.drop('pretest', axis=1).columns, 
                          data=rf_model.feature_importances_, 
                          columns=['features'])
```

```python
df_feat_rf = df_feat_rf[df_feat_rf['features']>=0.01]
```
