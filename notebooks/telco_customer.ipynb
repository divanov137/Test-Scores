```python
df = pd.get_dummies(df, columns=['gender', 'Partner', 'Dependents', 'PhoneService', 
                            'MultipleLines', 'InternetService', 'OnlineSecurity', 
                            'OnlineBackup', 'DeviceProtection', 'TechSupport',
                            'StreamingTV', 'StreamingMovies', 'Contract', 
                            'PaperlessBilling', 'PaymentMethod', 'Churn'], drop_first=True)
```

```python
df = df.drop(['customerID',
              'MultipleLines_No phone service',
              'OnlineSecurity_No internet service',
             'OnlineBackup_No internet service',
             'DeviceProtection_No internet service',
             'TechSupport_No internet service',
             'StreamingTV_No internet service',
             'StreamingMovies_No internet service'], axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
len(df), len(df.columns)
```

```python
df_corr = pd.DataFrame(df.corr()['Churn_Yes'])
```

```python
X = df.drop('Churn_Yes', axis=1)
```

```python
y = df['Churn_Yes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=124)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(X_train, y_train)
```

```python
y_pred = tree_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'criterion':['gini', 'entropy'],
             'max_depth':[2, 3, 4, 5, 6, 7, 8, 9],
              'min_samples_leaf':[0.01, 0.05, 0.1, 0.2, 0.3, 0.4],
             'max_leaf_nodes':[2, 4, 6, 8, 10, 12],
             'max_features':['auto', 'sqrt', 'log2']}
```

```python
simple_model = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(simple_model, parameters)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred_grid)
```

```python
print(classification_report(y_test, y_pred_grid))
```
